```python
import pandas as pd
import plotly.express as px
```

```python
# TODO: check why font size does not apply to the latex text

template = 'simple_white'
layout = {
    'paper_bgcolor': 'rgba(0,0,0,0)',
    'plot_bgcolor': 'rgba(0,0,0,0)',
    'font': {
        "family": "League Gothic",
        "size": 25,
        "color": "#586e75"
    }
}
```

```python
df = pd.read_csv ('solar_spectrum.csv', sep = ';', header = 1, names=['wavelength','AM0','gtilt','AM1.5'], index_col = 0)
df = df.drop (columns = 'gtilt')
df.head ()
```

```python
fig = px.line (
    df, 
    template = template
)
fig.update_layout (
    xaxis_title = r'$\lambda / \mathrm{nm}$',
    yaxis_title = r'$\textrm{Spektralno zračenje / }\mathrm{W m^{-2} nm^{-1}}$',
    legend_title = "Air Mass",
    ** layout
)
fig.show ()
fig.write_html ('solar_spectrum.html', include_mathjax = 'cdn')
```
